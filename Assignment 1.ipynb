# Data cleaning

```python
#Find out if we have missing rows
missing_rows = music_raw.isna().any(axis=1).sum()
missing_rows
```

```python
#Drop variables that don't correlate. 
music_df = music_raw.drop(['album_name', 'track_name', 'artists'], axis=1)
music_df
```

```python
# Select only the numerical columns for X
X_numeric = music_df.select_dtypes(include=[np.number])


# One-hot encode the categorical variables if needed
X_encoded = pd.get_dummies(music_df.select_dtypes(include=[object]))

# Combine numerical and encoded categorical data
music_df = pd.concat([X_numeric, X_encoded], axis=1)



music_df.head()
```

Dummy features is essential because most algorithms cannot handle categorical data in their raw form. By representing categories as separate binary features, we retain the important information that the categories represent without assuming any ordinal relationship between them. This is crucial for preserving the distinct characteristics of each category and ensuring they are appropriately factored into the model's predictions.

```python
# Step 2: Create 'popularity_category' for classification
music_df['popularity_category'] = pd.qcut(music_df['popularity'], q=3, labels=False)

# Step 3: Define features and target variables
X = music_df.drop(['popularity', 'popularity_category'], axis=1)  # Features for both tasks
y_regression = music_df['popularity']  # Target for regression
y_classification = music_df['popularity_category']  # Target for classification

# Perform a single train/test split on the selected features
X_train, X_test, y_train_reg, y_test_reg = train_test_split(X, y_regression, test_size=0.25, random_state=42)
X_train, X_test,y_train_class, y_test_class = train_test_split(X, y_classification, test_size=0.25, random_state=42)


print(music_df.head())
```

# Regression

R2 (R-squared) and MSE (Mean Squared Error) are two statistical metrics commonly used to evaluate the performance of regression models. R2 is a statistical measure that represents the proportion of the variance for a dependent variable that is explained by an independent variable or variables in a regression model. It is a value between 0 and 1, where 0 means that the model does not explain any of the t variability of the response data around its mean, and 1 means that it explains all the variability. MSE is a measure of the average of the squares of the errors. The error is the amount by which the value predicted by the model differs from the actual value. It is non-negative, and values closer to zero are better. A value of zero means that the model is a perfect predictor. 

Linear regression

```python
reg = linear_model.LinearRegression()
reg.fit(X_train, y_train_reg)
reg_pred = reg.predict(X_test)
print("Score on training set: {:.3f}".format(reg.score(X_train, y_train_reg)))
print("Score on test set: {:.3f}".format(reg.score(X_test, y_test_reg)))
```

```python
print("Regression MSE:", mean_squared_error(y_test_reg, reg_pred))
print("R2:", r2_score(y_test_reg, reg_pred))
```

The linear regression model has achieved very poor scores on both training and test.
The MSE for the linear regression model on the test set is approximately 373,2 which is a large number, indicating that the predictions are far off the actual.

# Classification

The F1 score and accuracy are two commonly used metrics to evaluate the performance of classification models. The F1 score is the harmonic mean of precision and recall. Precision is the number of true positive results divided by the number of all positive results, including those not identified correctly.F1 score reaches its best value at 1 (perfect precision and recall) and worst at 0. Accuracy is the ratio of correctly predicted observations to the total observations. It is the most intuitive performance measure and it is simply a ratio of correctly predicted observation to the total observations.
